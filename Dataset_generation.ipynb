```python
import os
import csv

def convert_to_csv(input_folder, output_file):
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Content', 'Class'])  # Writing header
        
        # Traverse through each folder
        for class_folder in os.listdir(input_folder):
            class_path = os.path.join(input_folder, class_folder)
            if os.path.isdir(class_path):
                class_name = class_folder
                
                # Traverse through each file in the class folder
                for file_name in os.listdir(class_path):
                    file_path = os.path.join(class_path, file_name)
                    if os.path.isfile(file_path):
                        with open(file_path, 'r', encoding='utf-8') as file:
                            content = file.read()
                            writer.writerow([content, class_name])

# Example usage:
input_folder = 'data'  # Update with the path to your folder containing class folders
output_file = 'output.csv'  # Name of the output CSV file
convert_to_csv(input_folder, output_file)
```

## Data from pdfs

```python
import os
import pandas as pd
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        if len(reader.pages)<40:
            for page_num in range(len(reader.pages)):
                text += reader.pages[page_num].extract_text()
        else:
            for page_num in range(40):
                text += reader.pages[page_num].extract_text()
    return text.strip()

def create_dataframe_from_pdfs(folder_path):
    data = {'Text': []}
    
    for filename in os.listdir(folder_path):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(folder_path, filename)
            text = extract_text_from_pdf(pdf_path)
            text = text.replace('\n', ' ')
            text = text.replace('\t', ' ')
            
            data['Text'].append(text)

            
    
    df = pd.DataFrame(data)
    return df
```

```python
import os
import pandas as pd
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        if len(reader.pages)<45:
            for page_num in range(len(reader.pages)):
                text += reader.pages[page_num].extract_text()
        elif len(reader.pages)>170:
            for page_num in range(59,len(reader.pages)):
                text += reader.pages[page_num].extract_text()
        else:
            for page_num in range(45):
                text += reader.pages[page_num].extract_text()
    return text.strip()

def create_dataframe_from_pdfs(folder_path):
    data = {'Text': [], 'Category': []}
    
    for filename in os.listdir(folder_path):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(folder_path, filename)
            text = extract_text_from_pdf(pdf_path)
            text = text.replace('\n', ' ')
            text = text.replace('\t', ' ')
            data['Text'].append(text)
            # data['Category'].append('Government')
    
    df = pd.DataFrame(data)
    return df
```

## Merged Dataset

```python
import pandas as pd
import os

# Path to the directory containing your CSV files
directory = 'Datasets'

dfs = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Read the CSV file into a DataFrame
        df = pd.read_csv(os.path.join(directory, filename))
        # Append the DataFrame to the list
        dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)

merged_df.to_csv('merged_data.csv', index=False)

print("Merged CSV file saved successfully.")
```
